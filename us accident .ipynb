```python
#df = pd.read_csv("./US_Accidents_May19_Migrated Data.csv", parse_dates=['Start_Time'])
df = pd.read_csv("./US_Accidents_May19_Migrated Data.csv")

df['Start_Time'] = pd.to_datetime(
    df['Start_Time'],
    format="%m/%d/%Y %I:%M:%S %p"
)
```

```python
df.Start_Time.head()
```

```python
df['Start_Time'] = pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time.dt.hour
```

```python
sns.histplot(df.Start_Time.dt.hour,bins=24,stat="proportion")
```

```python
df['day_name'] = df.Start_Time.dt.day_name()
```

```python
df.Start_Time.dt.day_name()
```

```python
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

sns.histplot(
    x=pd.Categorical(df.Start_Time.dt.day_name(), categories=day_order, ordered=True),
    stat="proportion"
)
```

```python
weekend_start_time = df.Start_Time[
    df.Start_Time.dt.day_name().isin(["Saturday", "Sunday"])
]
```

```python
sns.histplot(
    x=weekend_start_time.dt.hour,
    bins=24,
    stat="proportion"
)
```

```python
sns.histplot(df.Start_Time.dt.month,bins=12,stat="proportion")
```

```python
df.Start_Time.dt.year
```

```python
df_2017 = df[df.Start_Time.dt.year==2017]
```

```python
sns.histplot(
    x=df_2017.Start_Time.dt.month,
    bins=12,
    stat="proportion"
)
```

```python
df.Source.unique()
```

```python
df_2017_Bing = df_2017[df_2017.Source == 'Bing']
sns.histplot(
    x=df_2017_Bing.Start_Time.dt.month,
    bins=12,
    stat="proportion"
)
```

```python
df_2017_MapQuest = df_2017[df_2017.Source == 'MapQuest']
sns.histplot(
    x=df_2017_MapQuest.Start_Time.dt.month,
    bins=12,
    stat="proportion"
)
```

```python
df_2017_MapQuestBing = df_2017[df_2017.Source == 'MapQuest-Bing']
sns.histplot(
    x=df_2017_MapQuestBing.Start_Time.dt.month,
    bins=12,
    stat="proportion"
)
```
